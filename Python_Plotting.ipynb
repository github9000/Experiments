### Customizing seaborn graphics

Since seaborn is built on top of matplotlib, customization options for matplotlib will also work with seaborn. However, seaborn plotting functions often give much more scope for customization.

#### Choosing a colour palette

If you see the argument palette in the function, it means that you can choose a colour scheme. For example, here we choose a BuGn_r colour scheme, which is short for colours in the range Blue to Green (reversed) - that is, high values are blue and low values are blue. Here the hue is for wt, so the lightest cars will be green and the heaviest blue. See here (https://stanford.edu/~mwaskom/software/seaborn/tutorial/color_palettes.html) for a description of palettes available in seaborn.

Sometimes you see the argument cmap in a function instead of palette - this is the equivalent concept for Matplotlib functions. Here you can either use a Matplotlib colormap (http://matplotlib.org/users/colormaps.html) or convert a seaborn palette to the colormap format by simply casting the seaborn color_palette as a matplotlib ListedColorMap.

```python
from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.color_palette())
sns.color_palette('BuGn_r')
```

```python
ax = sns.swarmplot('gear', 'mpg', hue = 'wt',
                   size = 15, palette = "BuGn_r",
                   data = df)
ax.legend_.remove()
pass
```

```python
# if one wants a continuous colorbar/colormap, adding 256 as the number of colors required from Seaborn colorscheme helps.
cmap = ListedColormap(sns.color_palette("Spectral",256)) 
```

```python
# construct cmap
my_cmap = sns.light_palette("Navy", as_cmap=True)

N = 500
data1 = np.random.randn(N)
data2 = np.random.randn(N)
colors = np.linspace(0,1,N)
plt.scatter(data1, data2, c=colors, cmap=my_cmap)
plt.colorbar()
plt.show()
```

For complete control construct your own cmap using hex numbers

```python
from matplotlib.colors import ListedColormap

# construct cmap
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
my_cmap = ListedColormap(sns.color_palette(flatui).as_hex())

N = 500
data1 = np.random.randn(N)
data2 = np.random.randn(N)
colors = np.linspace(0,1,N)
plt.scatter(data1, data2, c=colors, cmap=my_cmap)
plt.colorbar()
plt.show()
```

### Laying Out Multiple Different Types of Plots

See http://matplotlib.org/users/gridspec.html for more layout exampes.

```python
plt.figure(figsize=(9,9))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2, 0))
ax5 = plt.subplot2grid((3,3), (2, 1))

sns.regplot('wt', 'mpg', data = df, ax = ax1)
sns.violinplot('gear', 'hp', data = df, ax = ax2)
sns.swarmplot('wt', 'gear', data = df, orient = "h",
              size = 10, alpha = 0.8, split = True, palette = 'pastel', ax = ax3)
sns.kdeplot(df.wt, df.hp, ax = ax4)
sns.barplot('cyl', 'hp', data = df,
            palette = sns.light_palette('orange'),  ax = ax5)
plt.tight_layout()
```

# Iris Dataset

PairGrid also allows you to quickly draw a grid of small subplots using the same plot type to visualize data in each. In a PairGrid, each row and column is assigned to a different variable, so the resulting plot shows each pairwise relationship in the dataset. This style of plot is sometimes called a scatterplot matrix, as this is the most common way to show each relationship, but PairGrid is not limited to scatterplots.

```python
iris = sns.load_dataset("iris")
g = sns.PairGrid(iris)
g.map(plt.scatter);
```

```python
# How many rows and columns are there in iris?
print(iris.shape)
# Display the first 6 rows.
iris.head()
```

Make a regression plot of Sepal.Length (x-coordinate) by Sepal.Width (y-coordinate) using the lmplot function. Add a title Bad regression.

```python
sns.lmplot(data=iris, x='sepal_length', y='sepal_width')
plt.title('Bad regression')
pass
```

Let's create a new figure where we fit separate linear regressions of Sepal.Length (x-coordinate) by Sepal.Width for each species on the same plot.

```python
sns.lmplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
pass
```

Now, we create a new figure where we have separate linear regression plots of sepal_length (x-coordinate) by sepal_width for each species. This figure should have 1 row and 3 columns.

```python
sns.lmplot(data=iris, x='sepal_length', y='sepal_width', col='species')
pass
```

We now create a new figure with 2 rows and 2 columns, where each figure shows a swarmplot comparing one of the 4 flower features (sepal_length, sepal_width, petal_length, petal_width) across species.

```python
data = pd.melt(iris, id_vars='species')
sns.factorplot(data=data, x='species', y='value', col='variable', col_wrap=2, kind='swarm')
pass
```

```python
# This time we set the titles for each plot
data = pd.melt(iris, id_vars='species')
g = sns.factorplot(data=data, x='species', y='value', col='variable', col_wrap=2, kind='swarm')
titles = ['One', 'Two', 'Three', 'Four']
for ax, title in zip(g.axes, titles):
    ax.set_title(title, fontsize=16)
pass
```

Saving the palette in a variable for later use in a plotting fucntion.
